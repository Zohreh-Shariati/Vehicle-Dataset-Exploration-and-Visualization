**Line Plot: City vs Highway MPG by Transmission Type**

```python
avg_mpg = df_car_cleaned.groupby('Transmission Type')[['city mpg', 'highway MPG']].mean()

plt.figure(figsize=(10,5))
sns.lineplot(data=avg_mpg, x='Transmission Type', y='city mpg', label='City MPG')
sns.lineplot(data=avg_mpg, x='Transmission Type', y='highway MPG', label='Highway MPG')
plt.title('Average City & Highway MPG by Transmission Type')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

Based on the line plot, Direct drive cars have a higher city mpg than highway mpg which shown a big peak in the line chart which is there is huge difference with other transmition type. on the left and right side of the peak, there are other type of transimision which are manual and automartic trnasmission , which both has higher average highway mpg than city mpg. and then there are automated manual, and unknown trnasimision (those cars which we dont have any information about their transmission type) , these both category distributed on the left and right side of the tail of the graph accordingly. both of them has a higher highway mpg compare to city mpg.

✅ How do MPG (city mpg and highway MPG) trends change with transmission type?
MPG trends change with transmission type as direct drive transmissions perform better in city driving, showing higher city MPG than highway MPG. In contrast, manual and automatic transmissions are more efficient on highways, with higher highway MPG compared to city MPG. Automated and unknown transmissions follow a similar trend, favoring highway driving. This indicates that transmission types are optimized for different driving conditions.

Direct drive transmissions have higher city MPG, indicating they are better suited for urban environments. Manual and automatic transmissions tend to have higher highway MPG, suggesting they are more efficient at sustained speeds. Automated and unknown transmissions also show higher highway MPG. Overall, transmission type strongly affects fuel efficiency, with city and highway driving preferences.

###**3.4 Correlation Analysis**
Investigate the correlation between the following variables:
*   Engine HP, MSRP, Popularity, city mpg, and highway MPG.




```python
corr_group= ['Engine HP', 'MSRP', 'Popularity', 'city mpg', 'highway MPG']
corr_matrix = df_car_cleaned[corr_group].corr()
print(corr_matrix)
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm_r',linewidths=0.5)
plt.title('Pearson Correlation Matrix')
plt.show()
```

**Correlation Analysis**

Engine HP & City MPG (-0.4, -0.38):
There is a moderate negative correlation between engine horsepower and city MPG (miles per gallon). Cars with more powerful engines typically have lower fuel efficiency in city driving. This is because high-horsepower engines require more fuel, particularly in stop-and-go traffic.

Engine HP & MSRP (0.65):
There is a moderate positive correlation between engine horsepower (HP) and the manufacturer's suggested retail price (MSRP). This means that cars with more horsepower generally tend to have a higher price.

MSRP & MPG (City/Highway) (-0.18, -0.19):
The correlation between MSRP and city/highway MPG is weak negative correlation. This suggests that higher-priced vehicles, often luxury or performance-oriented models, may prioritize power and comfort over fuel efficiency.

City MPG & Highway MPG (0.88):
There is a very strong positive correlation between city MPG and highway MPG. Cars that are efficient in the city are typically also efficient on the highway.

Popularity & Other variables (No correlation):
Popularity shows a near zero correlation with other variables, showing that factors like engine power, MSRP, or MPG have little impact on a car's popularity. Popularity might be influenced by other elements such as brand reputation, marketing, and consumer perceptions.
